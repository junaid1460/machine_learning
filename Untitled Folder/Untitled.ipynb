```python
import PIL.Image as image
import os, sys
import numpy as np
```

```python

#reading file
im = image.open('input.png')

im.show()
print(im.format, im.size, im.mode)
```

```python
def sub(im1, im2):
    r1, g1, b1 = im1.split()
    r2, g2, b2 = im2.split()
    r = np.array(r2) - np.array(r1)
    g = np.array(g2) - np.array(g1)
    b = np.array(b2) - np.array(b1)
    x = image.new(im1.mode, im1.size)
    return im.paste((r, g, b))
    
sub(im, im)
```

```python
im.rotate(10)
def scale(img, per):
    x, y = img.size
    x = int(x * (per / 100))
    y = int(y * (per / 100))
    return img.resize((x, y))
    
    
scale(im, 210)



    
```

```python
#saving image as jpg
im.save('out.jpg')
cropped = im.crop((10,10,100,100))
cropped
```

```python
#copying and pasting region of image
croppedx = cropped.transpose(image.ROTATE_180)

im.paste(croppedx, (10,10,100,100))

im.show()
```

```python

#swapping channels
r, g, b = im.split()
bgr = image.merge("RGB", (b, g, r))
bgr.show()
```

```python
source = im.split()


R, G, B = 0, 1, 2
# select regions where red is less than 100
mask = source[R].point(lambda i: i + 100)
# process the green band
out = source[G].point(lambda i: i + 100)
# out1 = source[B].point(lambda i: i + 100)
# paste the processed band back, but only where red was < 100
source[G].paste(out, None , mask)
# build a new multiband image
image.merge(im.mode, source)
```
