```python
import tensorflow as tf 
import numpy as np
from __future__ import print_function
import matplotlib.pyplot as plt
```

```python
def crange(f,t,i):
    while f < t:
        f+= i
        yield f
```

```python
x = [i for i in crange(0,10,0.01)]
y = [np.sin(i) for i in x]

plt.plot(x, y)
plt.show()
```
