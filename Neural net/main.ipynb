```python
def drange(_from,_till,_inc):
    if _from >= _till:
        return
    while _from < _till:
        yield _from
        _from += _inc
```

```python
array = [i for i in drange(-6, 6,0.01)]
```

```python
array
```

```python
import math

def f(x):
    return (1 - math.exp(-2 * x)/(1 + math.exp(-2 * x)))
        
output = []
for x in array:
    output += [f(x)]
output
    
```

```python
plt.plot(array,output)
```

```python
import numpy as np
def crop(img, x = 0, y = 0, width = None, height = None):
    out = []
    if width is None:
        width = len(img[0])
    if height is None:
        height = len(img)
    for i in range(height):
        out.append([])
        for j in range(width):
            out[i].append(img[i+y][j+x])
    return np.array(out)

def bitval(bits = []):
    string = ""
    for i in bits:
        if i >= 100:
            string += str(i)
        elif i >= 10:
            string += ("0" + str(i))
        else:
            string += ("00" + str(i))
            
    
    if len(bits) == 3:
        string += "255"
        
    return int(string)
def toonedim(array = []):
    out = []
    for i in array:
        for j in i:
            out.append(j)
    return out
def toBitnum(array = []):
    out = []
    for i in array:
        out += [ bitval(i)/255255255255]
    return out
            
        
            
```
